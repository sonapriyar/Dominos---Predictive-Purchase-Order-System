```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
pizza_sales = pd.read_csv(r"C:\Users\sonur\OneDrive\Desktop\Projects\Donimos\Raw data\Pizza_sales.csv")  # Replace with your actual file path
pizza_ingredients = pd.read_csv(r"C:\Users\sonur\OneDrive\Desktop\Projects\Donimos\Raw data\Pizza_ingrediants.csv")  # Replace with your actual file path

# Data Cleaning for pizza_sales
## Check for missing values
missing_values_sales = pizza_sales.isnull().sum()

## Handle missing values: Drop rows with missing values
pizza_sales.dropna(inplace=True)

## Check for duplicates
duplicates_sales = pizza_sales.duplicated().sum()

## Remove duplicates
pizza_sales.drop_duplicates(inplace=True)

# Handle outliers for total_price
def remove_outliers(df, column):
    z_scores = np.abs((df[column] - df[column].mean()) / df[column].std())
    return df[z_scores < 3]  # Keep rows where z-score < 3

# Set the sales column to 'total_price'
sales_column = 'total_price'
pizza_sales = remove_outliers(pizza_sales, sales_column)

# Data Formatting for pizza_sales
# Convert date columns to datetime format
pizza_sales['order_date'] = pd.to_datetime(pizza_sales['order_date'])

# Data Cleaning for pizza_ingredients
## Check for missing values
missing_values_ingredients = pizza_ingredients.isnull().sum()

## Handle missing values: Leave blanks as is and drop rows with all NaN values
pizza_ingredients = pizza_ingredients.dropna(how='all')

## Check for duplicates
duplicates_ingredients = pizza_ingredients.duplicated().sum()

## Remove duplicates
pizza_ingredients.drop_duplicates(inplace=True)

# Example column to check for blanks (assuming it has a column named 'ingredient_name')
ingredient_column = 'ingredient_name'  # Update with actual ingredient column name
if ingredient_column in pizza_ingredients.columns:
    blanks_count = pizza_ingredients[ingredient_column].isnull().sum()  # Count blanks
else:
    blanks_count = None

# EDA for pizza_sales
## Descriptive statistics
descriptive_stats = pizza_sales.describe()

## Sales trend over time
plt.figure(figsize=(12, 6))
pizza_sales.groupby('order_date')[sales_column].sum().plot()
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid()
plt.savefig('total_sales_over_time.png')  # Save the plot as a PNG file
plt.close()  # Close the plot to avoid display

## Sales distribution
plt.figure(figsize=(12, 6))
sns.histplot(pizza_sales[sales_column], bins=30, kde=True)
plt.title('Sales Distribution')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.grid()
plt.savefig('sales_distribution.png')  # Save the plot as a PNG file
plt.close()  # Close the plot to avoid display

## Correlation heatmap
# Select only numeric columns for correlation
numeric_cols = pizza_sales.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(12, 8))
sns.heatmap(pizza_sales[numeric_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.savefig('correlation_heatmap.png')  # Save the heatmap as a PNG file
plt.close()  # Close the plot to avoid display

# EDA for pizza_ingredients
## Check unique ingredients (assuming it has a column named 'ingredient_name')
if ingredient_column in pizza_ingredients.columns:
    unique_ingredients = pizza_ingredients[ingredient_column].nunique()
else:
    unique_ingredients = None

# Prepare EDA summary
eda_summary = {
    'Metric': ['Missing Values', 'Duplicates', 'Unique Ingredients', 'Blanks Count'],
    'Values': [missing_values_sales.to_dict(), duplicates_sales, unique_ingredients, blanks_count],
}

# Create a DataFrame for EDA summary
eda_summary_df = pd.DataFrame(eda_summary)

# Save EDA summary to a CSV file
eda_summary_df.to_csv('eda_summary.csv', index=False)

# Save missing values information for ingredients to a CSV file
missing_values_ingredients.to_csv('missing_values_ingredients.csv', index=True)

# Final Data Overview
print("\nData overview after cleaning:")
print("Pizza Sales Data:")
print(pizza_sales.info())
print("\nPizza Ingredients Data:")
print(pizza_ingredients.info())
```
